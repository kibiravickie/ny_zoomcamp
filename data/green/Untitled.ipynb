```python
import os
```

```python
years = [2019, 2020]
months = range(1, 13)
```

```python
for year in years:
    for month in months:
        url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_{year}-{month}.csv.gz"
        file_name =f"yellow_tripdata_{year}-{month:02}"
        os.system(f"wget {url} -O {file_name}")
```
